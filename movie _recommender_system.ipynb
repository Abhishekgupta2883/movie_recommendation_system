```python
import numpy as np
import pandas as pd

# Load the datasets
credits = pd.read_csv('/content/tmdb_5000_credits.csv')
movies = pd.read_csv('/content/tmdb_5000_movies.csv')
```

```python
# Display the first 5 rows of the credits DataFrame
display(credits.head(1))

# Display the first 5 rows of the movies DataFrame
display(movies.head(1))
```

```python
movies = movies.merge(credits, on='title')
movies.shape
```

```python
movies.info()
```

```python
columns_to_keep = [
    'genres',
    'id',
    'keywords',
    'popularity',
    'release_date',
    'overview',
    'cast',
    'crew',
    'vote_average',
    'vote_count',
    'title'
]

# Get all columns currently in the DataFrame
all_columns = movies.columns.tolist()

# Identify columns to drop
columns_to_drop = [col for col in all_columns if col not in columns_to_keep]

# Drop the identified columns
movies = movies.drop(columns=columns_to_drop)

# Display the first 5 rows and the new shape of the DataFrame
display(movies.head(1))
print(movies.shape)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
import ast
def convert(obj):
  L =[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies["genres"]= movies["genres"].apply(convert)
```

```python
movies["keywords"]= movies["keywords"].apply(convert)
```

```python
movies["crew"][0]
```

```python
def convert3(obj):
  L =[]
  counter =0
  for i in ast.literal_eval(obj):
    if counter >=3: break
    L.append(i['name'])
    counter+=1
  return L
```

```python
movies["cast"]= movies["cast"].apply(convert3)
```

```python
def convert1(obj):
  L =[]
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L
```

```python
movies["crew"]= movies["crew"].apply(convert1)
```

```python
movies["overview"]= movies["overview"].apply(lambda x:x.split())
```

```python
movies.head(1)
```

```python
movies["crew"]= movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
movies["genres"]= movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"]= movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"]= movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies["tags"] = movies['overview']+movies['cast']+movies['crew']+movies['genres']+movies['keywords']
```

```python
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x))
```

```python
from datetime import datetime

def convert_date_to_seconds(date_str):

    try:
        # Ensure date_str is a string before stripping and parsing
        date_str = str(date_str).strip()
        # Parse the date string into a datetime object
        date_obj = datetime.strptime(date_str, '%Y-%m-%d')
        # Return the timestamp (seconds since epoch) as an integer
        return int(date_obj.timestamp())
    except (ValueError, TypeError):
        # Handle cases where date_str is not a valid date string or is already a non-string type
        print(f"Warning: Could not parse date string '{date_str}'. Returning 0.")
        return 0

movies['release_date'] = movies['release_date'].apply(convert_date_to_seconds)

# Display the first few rows to check the conversion
display(movies.head(1))

# Display info to confirm the data type change
movies.info()
```

```python
movies.info()
```

```python
df = movies[['id','title','tags','vote_average','popularity','release_date']].copy()
df.head(1)
```

```python
df['tags'] = df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer

# Initialize the Porter Stemmer
ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

# Apply stemming to the 'tags' column
df['tags'] = df['tags'].apply(stem)

# Display the first few rows to check the stemmed tags
display(df.head(1))
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Initialize CountVectorizer with specified parameters
cv = CountVectorizer(max_features=5000, stop_words='english')

# Fit and transform the 'tags' column
vectors = cv.fit_transform(df['tags']).toarray()

# Display the shape of the resulting vectors array
print(vectors.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)

# Display the shape of the similarity matrix and its first few rows
print(similarity.shape)
display(similarity)
```

### Incorporating Numerical Features into Recommendations

To effectively use `vote_average`, `popularity`, and `release_date` for recommendations, we first need to normalize them. This ensures that features with larger values (like popularity) don't disproportionately influence the similarity calculation compared to features with smaller scales (like vote average). We will use `MinMaxScaler` to scale these features to a range between 0 and 1.

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Select numerical features for scaling
numerical_features = ['vote_average', 'popularity', 'release_date']

# Apply scaling to the selected features in a copy of the DataFrame
df_scaled = df[numerical_features].copy()
df_scaled[numerical_features] = scaler.fit_transform(df_scaled[numerical_features])

# Display the first few rows of the scaled DataFrame
display(df_scaled.head())
```

Now that our numerical features are scaled, we can calculate a new similarity based on these features. We will then combine this with our existing tag-based similarity using a weighted average. This will allow both content (tags) and film metrics (`vote_average`, `popularity`, `release_date`) to contribute to the recommendation score.

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate cosine similarity for numerical features
numerical_similarity = cosine_similarity(df_scaled)

# Define weights for the two types of similarity
# You can adjust these weights based on how much you want each to influence the recommendation
weight_tags = 0.7
weight_numerical = 0.3

# Combine the existing tag-based similarity with the new numerical similarity
# Ensure both similarity matrices have the same shape for element-wise multiplication
combined_similarity = (weight_tags * similarity) + (weight_numerical * numerical_similarity)

# Display the shape and a snippet of the combined similarity matrix
print('Shape of combined similarity matrix:', combined_similarity.shape)
display(combined_similarity[:5, :5])
```

Finally, we will update our `recommend` function to use this `combined_similarity` matrix.

```python
def recommend_enhanced(movie):
    movie_index = df[df['title'] == movie].index[0]
    distances = combined_similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    print(f"Top 5 recommendations for '{movie}' (enhanced):")
    for i in movies_list:
        print(df.iloc[i[0]].title)

# Example usage of the enhanced recommendation function
recommend_enhanced('Avatar')
recommend_enhanced('The Dark Knight Rises')
```
